```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Experiment 1
- Input Data Type: Integer
- Random Number Upperbound: 10000
- Maximum Number of Inputs: 100

```python
x1= pd.read_csv('NumOfInputs-n_100.csv')
x1
```

```python
y1= pd.read_csv('NumOfOperations_int-100.csv')
y1
```

```python
# Data Type = Integer
# Upperbound Value = 10000
# Number of Maximum Inputs = 100
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")
plt.ylabel("Number of operation")
plt.plot(x1.to_numpy(),y1.to_numpy())
plt.show()
```

### Experiment 4
- Input Data Type: Floating Point
- Random Number Upperbound: 10000

```python
x= pd.read_csv('NumOfInputs-n_100.csv')
x
```

```python
y= pd.read_csv('NumOfOperations_fp-100.csv')
y
```

```python
# Data Type = Floating Point
# Upperbound Value = 10000
# Number of Maximum Inputs = 100
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")
plt.ylabel("Number of operation")
plt.plot(x.to_numpy(),y.to_numpy())
plt.show()
```

```python
# Show the comparison between high and low upperbound
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")

plt.ylabel("Number of operation")
plt.plot(x2.to_numpy(), y2.to_numpy(), label = "Upperbound = 10000")    # high upperbound
plt.plot(x3.to_numpy(), y3.to_numpy(), label = "Upperbound = 100")   # low upperbound
plt.legend()

plt.show()
```
