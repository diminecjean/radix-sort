## Relationship Analysis

### Relationship between high and low upper bound on the number of operations

```python
# Show the comparison between high and low upperbound
# Data type: Integer
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")
plt.ylabel("Number of operation")

plt.plot(x2.to_numpy(), y2.to_numpy(), label = "Int:Upperbound = 10000") 
plt.plot(x3.to_numpy(), y3.to_numpy(), label = "Int:Upperbound = 100")   
plt.legend()

plt.show()
```

```python
# Show the comparison between high and low upperbound
# Data type: Floating Point
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")
plt.ylabel("Number of operation")

plt.plot(x5.to_numpy(), y5.to_numpy(), label = "FP:Upperbound = 10000") 
plt.plot(x6.to_numpy(), y6.to_numpy(), label = "FP:Upperbound = 100")   
plt.legend()

plt.show()
```

From the graphs above, we can infer that the size of the upperbound, which indicates the maximum value of the randomly generated inputs, only affects the number of operations for inputs of integer data type, but not inputs of floating point data type. This is because the number of iterations in the radix sort algorithm for integers is based on the number of digits of the integer, while the number of iterations for floating points is based on the number of digits of the whole number, including both the integer places and the decimal places. For integer values, the higher the upperbond, the larger the number of operations performed.

Below is a graph that illustrates the relationship for both integers and floating points.

```python
# Show the comparison between high and low upperbound
# Data type: Integer and Floating Point
plt.title("Number of operation vs n.")
plt.xlabel("Number of input, n")
plt.ylabel("Number of operation")

plt.plot(x2.to_numpy(), y2.to_numpy(), label = "Int:Upperbound = 10000") # integer
plt.plot(x3.to_numpy(), y3.to_numpy(), label = "Int:Upperbound = 100")   # integer
plt.plot(x5.to_numpy(), y5.to_numpy(), label = "FP:Upperbound = 10000") # floating point
plt.plot(x6.to_numpy(), y6.to_numpy(), label = "FP:Upperbound = 100")   # floating point
plt.legend()

plt.show()
```
